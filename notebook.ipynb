```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('house price prediction.csv')
```

```python
df
```

```python
df.plot(kind="scatter",x="Area",y="Price",color="red",marker="+")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.plot()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
